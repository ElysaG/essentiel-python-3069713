Créer classe Python nommée CompteBancaire qui représente un compte bancaire avec les attributs: 
- numeroCompte
- nom
- solde

```python
class CompteBancaire:
  def __init__(self,numeroCompte,nom,solde):
    self.numeroCompte=numeroCompte
    self.nom=nom
    self.solde=solde
  
  def infos(self):
    return print(f"Votre banque est le {self.nom} votre solde est de de {self.solde}")
  def versement(self,montantVerst):
    self.solde +=montantVerst
    return print(f"Après versement de {montantVerst} votre solde est de de {self.solde}")
  
  def retrait(self,montantRetrait):
    if montantRetrait <= self.solde :
      self.solde -=montantRetrait
      return print(f"Après retrait de {montantRetrait} votre solde est de de {self.solde}")
    else:
       return print(f"Votre solde est de {self.solde} et ne vous permet pas d'effectuer un retrait")
  
  def agios(self):
    self.solde = self.solde*95/100

  def afficher(self):
    print("Compte numéro:",self.numeroCompte)
    print("Nom Banque:",self.nom)
    print("Solde:",self.solde)
```

```python
moncompte = CompteBancaire("35432432","LCL",10000)
```

```python
moncompte.infos()
```

```python
moncompte.retrait(20)
```

```python
moncompte.afficher()
```
