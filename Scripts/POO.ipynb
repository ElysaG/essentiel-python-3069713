```python
class Salarie(object): # Explicite (ancienne syntaxe)
  def __init__(self, nom, nos):
    self.nom= nom
    self.nos=nos
  
  def afficher (self):
    print("Nom:", self.nom)
    print("Num salarié:", self.nos)
```

```python
lisa = Salarie("Léa","03626552525")
```

```python
lisa.afficher()
```

```python
class Manager(Salarie):
  def __init__(self,nom,nos,equipe):
    super().__init__(nom,nos)
    self.equipe=equipe

  def afficher_equipe(self):
    print("Equipe:",self.equipe)

    
```

```python
employe = Salarie("Alice","S011")
manager=Manager("Bob","MI252","IT")
```

```python
employe.afficher()
```

```python
manager.afficher_equipe()
```

```python
# Classe de base
class Salarie:
    def __init__(self, nom, nos):
        self.nom = nom
        self.nos = nos
    
    def afficher(self):
        print(f"Nom: {self.nom}, N°: {self.nos}")

# Classe enfant 1
class Developpeur(Salarie):
    def __init__(self, nom, nos, langage):
        super().__init__(nom, nos)
        self.langage = langage
    
    def coder(self):
        print(f"{self.nom} code en {self.langage}")

# Classe enfant 2
class Manager(Salarie):
    def __init__(self, nom, nos, equipe):
        super().__init__(nom, nos)
        self.equipe = equipe
    
    def gerer(self):
        print(f"{self.nom} gère l'équipe {self.equipe}")

# Utilisation
dev = Developpeur("Alice", "D001", "Python")
mgr = Manager("Bob", "M001", "IT")

dev.afficher()   # Méthode HÉRITÉE
dev.coder()      # Méthode PROPRE

mgr.afficher()   # Méthode HÉRITÉE
mgr.gerer()      # Méthode PROPRE
```

```python
class Vehicule:
  def __init__(self):
    self.est_allume=True
    self.vitesse =10
  
  def obtenir_vitesse(self):
    return self.vitesse
  def deplacement(self):
    if self.est_allume and self.vitesse>0:
      print("le véhicule se déplace")
```

```python
v=Vehicule()
```

```python
v.deplacement()
```

```python
class VehiculeMaritime(Vehicule):
  def naviguer(self):
    print("Le véhicule maritime navigue")
  def deplacement(self):
    if self.est_allume and self.obtenir_vitesse()>0:
      self.naviguer()
```

```python
vm = VehiculeMaritime()
```

```python
vm.deplacement()
```
