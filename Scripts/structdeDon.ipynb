```python

def calcul_rem(temps,Sbase=50):
  if temps <= 40:
    Rem = temps * Sbase
  else:
    Rem = 40 *Sbase +(temps-40)*Sbase*1.5
  return Rem
```

```python
calcul_rem(50)
```

```python
def Salaire(heures_travaillees,tarif_horaire):
  if heures_travaillees>40:
    return tarif_horaire*40+(heures_travaillees-40)*1.5
  else:
    return heures_travaillees*tarif_horaire
```

```python
heures_travaillees=50
tarif_horaire=100
```

```python
paie=Salaire(heures_travaillees,tarif_horaire)
paie
```

```python
print(f"Total à payer : {paie:.2f}","€")
```
